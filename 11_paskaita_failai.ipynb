```python
with open("paskaitajokes.txt", "r") as jokes_file:
    joke, full_joke = "", ""
    while True:
        fragment = jokes_file.read(2)
        pozicija = jokes_file.tell()
        if not fragment:
            break
        joke += fragment
        if joke.find(".") > 0:
            full_joke = joke[:joke.find(".")+1]
            joke = joke[joke.find(".")+1:]
        if len(full_joke):
            print(full_joke)
        full_joke = ""
```

```python
with open("paskaitajokes.txt", "r") as jokes_file:
    while True:
        fragment = jokes_file.read(8)
        pozicija = jokes_file.tell()
        if not fragment:
            break
        joke += fragment
        jokes_file.seek(pozicija - 2)
        if len(fragment) <= 6:
            break
    print(joke)
```
